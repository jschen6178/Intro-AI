```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
x = np.array([0,1,2,3])
y = np.array([0,0.45,1.1,1.55])
weights = np.array([0, 0.5, 1, 1.5])
for weight in weights:
    ys = weight*x
    plt.plot(x,ys)
    
plt.scatter(x,y)
plt.show()

```

```python
def costFunction(xvalues, yvalues, weight):
    predictions = weight*xvalues
    costArr = []
    for i in range(len(predictions)):
        costArr.append((predictions[i]-yvalues[i])**2)
    cost = sum(costArr)/(2*len(predictions))
    return cost
```

```python
costs = np.zeros(len(weights))

for i in range(len(weights)):
    costs[i] = costFunction(x, y, weights[i])
print(costs)
plt.plot(weights,costs)
plt.show()
```

```python
def derivCost(x, y, weight):
    predictions = weight*x
    costArr = []
    for i in range(len(predictions)):
        costArr.append((predictions[i]-y[i])*x[i])
    derivative = sum(costArr)/len(predictions)
    return derivative

derivatives = np.zeros(len(weights))

for i in range(len(weights)):
    derivatives[i] = derivCost(x, y, weights[i])
print(derivatives)
plt.plot(weights,derivatives)
plt.show()

alpha = 0.3
```
