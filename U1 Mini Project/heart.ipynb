```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
from sklearn import neighbors
from sklearn.model_selection import train_test_split
df = pd.read_csv('heart.csv', engine = 'python')
#getting the features, age, blood pressure, cholesterol
age = df.iloc[:, 0]
rbpm = df.iloc[:, 3]
chol = df.iloc[:, 4]
#putting the features into one, usable list
x = []
for i in range(len(age)):
    x.append([age[i], rbpm[i], chol[i]])
#labels here, high blood sugar, yes is 1, no is 0.
y = df.iloc[:, 5]
```

```python
#setting up the model, splitting datasets, getting KNN
x_train, x_test, y_train, y_test = train_test_split(x, y, train_size=0.9, )
knn = neighbors.KNeighborsClassifier(n_neighbors=3)
knn.fit(x_train, y_train)
# Compute the score (mean accuracy) on test set
score = knn.score(x_test, y_test)
print('KNN score: %f' % score)
```

```python
#getting the actual predictions from the x_test dataset
predictions = knn.predict(x_test)
#getting the individual data sets in each x_test, age blood rpessure and chol
age_x_test = [i[0] for i in x_test]
rbpm_x_test = [i[1] for i in x_test]
chol_x_test = [i[2] for i in x_test]

#scatter plot 1
plt.scatter(age_x_test, rbpm_x_test, c=predictions, alpha = 0.6)
plt.xlabel("Age (years)")
plt.ylabel("Rest Blood Pressure (mm of Hg)")
plt.title("Age vs Blood Pressure")
plt.show

```

```python
#scatter plot 2
plt.scatter(age_x_test, chol_x_test, c=predictions, alpha = 0.6)
plt.xlabel("Age (years)")
plt.ylabel("Cholesterol (mg/dl)")
plt.title("Age vs Cholesterol")
plt.show
```

```python
#scatter plot 3
plt.scatter(chol_x_test, rbpm_x_test, c=predictions, alpha = 0.6)
plt.xlabel("Cholesterol (mg/dl)")
plt.ylabel("Rest Blood Pressure (mm of Hg)")
plt.title("Cholesterol vs Blood Pressure")
plt.show
```
